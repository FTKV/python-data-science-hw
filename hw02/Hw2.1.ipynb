## Частина перша: Знайомство з Pandas.
Прочитайте дані за допомогою методу read_html з таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)" посилання

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
URL="https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8"
TABLE_NAME = "Коефіцієнт народжуваності в регіонах України"
data = pd.read_html(URL, match=TABLE_NAME, decimal=",", thousands=".")[0]
```

Необхідно виконати:

Вивести перші рядки таблиці за допомогою методу head

```python
data.head()
```

Визначте кількість рядків та стовпців у датафреймі (атрибут shape)

```python
data.shape
```

Замініть у таблиці значення "—" на значення NaN

```python
data.replace("—", np.nan, inplace=True)
```

Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
data.dtypes
```

Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
data["2014"] = pd.to_numeric(data["2014"])
data["2019"] = pd.to_numeric(data["2019"])
```

```python
data.dtypes
```

Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
data.isnull().sum()/len(data)
```

```python
plt.xlabel('Рік', fontsize='small', color='midnightblue')
plt.ylabel('Коефіцієнт народжуваності', fontsize='small', color='midnightblue')
plt.title('Коефіцієнт народжуваності в Україні', fontsize=15)
plt.ylim([0, 25])
plt.plot(data.columns[1:], data[data["Регіон"] == "Україна"].values[0][1:])
plt.show()
```

Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
data.drop(index=data.index[-1], inplace=True)
data
```

Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
for col_name in list(data):
    if col_name!="Регіон":
        data[col_name] = data[col_name].fillna((data[col_name].mean()))
data
```

Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
data[data["2019"] > data["2019"].mean()][["Регіон"]]
```

У якому регіоні була найвища народжуваність у 2014 році?

```python
data[data["2014"] == data["2014"].max()][["Регіон"]]
```

Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
plot = pd.DataFrame({"2019": data["2019"].values}, index=data["Регіон"]).plot(kind='bar', title=TABLE_NAME + " у 2019 році")
```

```python
width = 18
height = 8
sns.set(rc = {'figure.figsize':(width, height)})
strp_plot = sns.swarmplot(data=data).set(title="Коефіцієнт народжуваності в регіонах України", xlabel='Рік', ylabel='Коефіцієнт народжуваності')
```
